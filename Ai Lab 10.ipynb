<b>4. Data Pre-Preprocessing<b>

**4.1 Manage null values**

```python
df.isnull().sum()
```

```python
# object
def fillNaObjMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])
coulmns = ["street", "city", "statezip"]
fillNaObjMode(coulmns)
```

```python
df.info()
```

```python
# Integer
def fillNaIntMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])
columns = ['bedrooms','bathrooms','floors','waterfront','view','yr_built','yr_renovated']
fillNaIntMode(coulmns)
```

```python
# Float
def fillNaFloatMean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())
columns = ['price','condition','sqft_living','sqft_lot','sqft_above','sqft_basement']
fillNaFloatMean(coulmns)
```

```python
df.info()
```

**4.2 Manage datatypes**

```python
df["price"] = df["price"].astype("int64")
```

```python
# For Float Type
def changeFloatToInt(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['price','condition','sqft_living','sqft_lot','sqft_above',
           'sqft_basement','bedrooms','bathrooms','floors','waterfront',
           'view','yr_built','yr_renovated']
changeFloatToInt(columns)
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))    # data_LE.fit(data)
        df[i] = data_LE.transform(df[i])
        
columns = ['street', 'city', 'statezip']
convertObjToInt(columns)
```

```python
df.info()
```

```python
data = pd.DataFrame({"city":df["city"].unique()})
data_LE = LabelEncoder()
data_LE.fit(np.ravel(data))
# data_LE.fit(data)
df["city"] = data_LE.transform(df["city"])
```

```python
def changeObjectToInt(cols):
    for i in cols:
        data = pd.DataFrame({"i":df["i"].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))
        df["i"] = data_LE.transform(df["i"])
coulmns = ["street", "city", "statezip"]
changeObjectToInt(coulmns)
```

```python
df.to_csv("Processed_data.csv", index = False, header = True)
```
