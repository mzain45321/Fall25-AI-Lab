```python
# df.drop(columns='date',inplace=True)  
# df.drop('date', axis=1,inplace=True)
df = df.drop('date', axis=1)
```

```python
df = df.drop('country', axis=1)
```

```python
# for object or string
df['street'].mode()[0]   # most frequent value
```

```python
# df['street'].fillna(df['street'].mode()[0])
df['street'] = df['street'].fillna(df['street'].mode()[0])
```

```python
def fillnaobjmode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])
columns = ['street','city','statezip']
fillnaobjmode(columns)
```

```python
# for float values
def fillnafloatmean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())
columns = ['price','sqft_living','sqft_lot','condition','sqft_above','sqft_basement']
fillnafloatmean(columns)
```

```python
# for int values
def fillnaintmode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])
columns = ['bedrooms','bathrooms','waterfront','floors','view','yr_built','yr_renovated']
fillnaintmode(columns)
```

```python
df['price'] = df['price'].astype('int64')
```

```python

def changefloat(cols):
    for i in cols:
        df[i] = df[i].astype('int64')
columns = ['price','sqft_living','sqft_lot','condition',
           'sqft_above','sqft_basement','bedrooms','bathrooms','waterfront',
           'floors','view','yr_built','yr_renovated']
changefloat(columns)
    
```

```python
data = pd.DataFrame({'city':df['city'].unique()})
data_LE = LabelEncoder()
data_LE.fit(np.ravel(data))
```

```python
def changeobjtoint(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))
        # data_LE.fit(data)
        df[i] = data_LE.transform(df[i])
columns = ['street','city','statezip']
changeobjtoint(columns)
    
```

```python
df.to_csv('preprocessed.csv',index= False,header = True)
```

```python
df = pd.read_csv("preprocessed.csv")
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('price',axis=1) # with resultant column
y = df['price']    # with resultant column
```

```python
x.head(1)
```

```python
y.head(1)
```

```python
trainx,testx,trainy,testy = train_test_split(x,y,test_size=0.3,shuffle=False) # 70 percent training and 30 percent testing
```

```python
trainx
```

```python
print(trainx.shape,trainy.shape)
print(testx.shape,testy.shape)
```

**Apply classifier/ model training**

```python
from sklearn.svm import SVC
modelsvc = SVC()
modelsvc.fit(trainx,trainy)
print(modelsvc)
```

```python
import pickle
pickle.dump(modelsvc,open('modelsvc.pkl','wb'))
```

```python
#loading model
modelsvc = pickle.load(open('modelsvc.pkl','rb'))
```

```python
modelsvcpred = modelsvc.predict(testx)
modelsvcpred
```

```python
from sklearn.metrics import accuracy_score
modelsvcaccu = accuracy_score(testy,modelsvcpred)
print(round(modelsvcaccu,3)*100,'%')
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

modelsvcpred = modelsvc.predict(testx)

accuracy_svc = round(accuracy_score(testy, modelsvcpred) * 100, 2)
print(f"SVC Accuracy: {accuracy_svc} %\n")

print("Classification Report:")
print(classification_report(testy, modelsvcpred))

print("Confusion Matrix:")
print(confusion_matrix(testy, modelsvcpred))

comparison_svc = pd.DataFrame({
    'Actual': testy,
    'Predicted': modelsvcpred,
    'Correct': testy == modelsvcpred
})

print("\nComparison Table (first 10 rows):")
print(comparison_svc.head(10))
```
